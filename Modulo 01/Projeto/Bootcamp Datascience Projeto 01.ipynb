#### **3.1 Importação dos Pacotes Utilizadas**

# **4. Análise Exploratória dos Dados**

Análise Exploratória dos Dados é a etapa em que utilizaremos técnicas para a exploração dos dados através de tabelas e gráficos disponíveis nas bibliotecas importadas anteriormente. Com isso, podemos observá-los e formular hipóteses e extrair insights.

###**4.1 Análise Exploratória por quantidade de internações por Estado**

````python
dados['Total'].sum()
````

Somando os Totais com função sum() do pandas, verificamos que tivemos **149 milhões de internações** no SUS no período de **Janeiro de 2008 a Março de 2021**.

Plotamos em um gráfico de barra para compararmos a quantidade de internações por estado.

````python
ax = dados.plot(x="Unidade da Federação", y="Total", kind="bar", figsize=(9,6), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Número de internações por unidade da federação")
plt.show()
````

Ao analisar o gráfico acima, podemos concluir que **São Paulo** está em **1º lugar** em torno de **30.000.000 internações**, **Minas Gerais em 2º lugar** em torno **15.000.000** internações e **Bahia  em 3º lugar** em torno de **10.000.000 internações**. Ou seja, São Paulo teve o **dobro de internações** comparado ao 2º lugar no mês em questão.

Porém, com este gráfico desordenado fica difícil analisar outros estados e sua classificação no número de internações no SUS, por exemplo: **Acre, Roraima e Amapá**. Então utilizaremos a funcão **sort_values** do pandas para ordená-lo e facilitar essa análise.

````python
ax = dados.sort_values('Total', ascending=False).plot(x='Unidade da Federação', y='Total', kind="bar", figsize=(9,6), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Número de internações por unidade da federação")
plt.show()
````

Pronto! Ordenando os gráficos, podemos visualizar a classificação de cada estado. Por exemplo: **Roraima, Amapá e Acre** são os estados com os menores números de internações.

A partir desse gráfico já podemos tirar algumas conclusões:

#### Conclusão 1: Região Sul e Sudeste prevalece com 6 estados (São Paulo, Minas Gerais, Paraná, Rio Grande do Sul e Santa Catarina) no top 10 de números de internações.

#### Conclusão 2: Pará é o único estado da região Norte no top 10 de internações.

#### Conclusão 3: Região Norte prevalece como o estado com os menores números de internações.

###**4.2 Análise Exploratória por quantidade de internações por Mês**

Neste momento, vamos analisar a quantidade de **internações por mês** e para isso precisaremos excluir a coluna Total dos dados para não distorcer o gráfico.

Utilizaremos a função **drop** do pandas para realizar essa exclusão.

````python
dados = dados.drop("Total", axis=1)
````

Para plotar o gráfico, precisaremos fazer dois tratamentos em nossos dados antes:

**1º: Precisamos tratar os dados faltantes.**

**2º: Precisamos definir coluna Unidade da Federação como index.**

#### 4.2.1 Tratando dados faltantes

De acordo com a documentação do [DATASUS](http://www2.datasus.gov.br/DATASUS/index.php?area=0202&id=11633&VObj=http://tabnet.datasus.gov.br/cgi/deftohtm.exe?sih/cnv/qi), as colunas com os números de internações que não foram preenchidas são marcadas com um traço ("-"). 

Então é necessário usar a função **replace**, para substituir os campos preenchidos com traço por valores nulos.

````python
dados = dados.replace('-',np.nan)
````

Com isso, podemos utilizar a função do pandas **isnull** e contabilizarmos a quantidade desses valores e decidirmos a melhor maneira para o tratamento dos valores faltantes, 

````python
dados.isnull().sum().sum()
````

Ao contabilizar os valores faltantes com a função **sum** e **isnull**, verificamos que temos apenas um registro. Então, como é apenas um valor, posso excluí-lo com a funçao **dropna** sem grandes impactos na análise.

````python
dados = dados.dropna()
````

````python
dados.isnull().sum().sum()
````

Valor excluído com sucesso!

#### 4.2.2 Definir coluna Unidade da Federação como index.

Precisamos definir a coluna Unidade da Federação como **index**, pois ela é do tipo texto, e quando plotamos um gráfico com funcao **plot** do pandas, ele vai tentar somar todos os valores das colunas fora do **index**, inclusive a Unidade da Federação, então como este campo não é um número, e sim, um texto, ocorrerá um erro no pandas.

````python
dados = dados.set_index('Unidade da Federação')
````

````python
dados.head()
````

#### Para o plot, utilizamos a função astype do pandas para converter os valores quantitativos de internações e garantir que eles sejam do tipo numérico.

````python
dados = dados.astype('float64')
````

````python
ax = dados.T.plot(figsize=(10,7.5), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.legend(bbox_to_anchor = (1, 1))
plt.show()
````

Observando o gráfico para analisar as linhas de tendências por período, percebemos o quão difícil é analisá-lo devido a quantidade de Estados.

Para melhor a visualização para ajudar na análise, podemos selecionar os 6 estados.

````python
ax = dados[:6].T.plot(figsize=(10,7.5), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.legend(bbox_to_anchor = (1, 1))
plt.title("Os Estados do Norte por números de internações")
plt.show()
````

Agora sim! Selecionando 6 estados e plotando no gráfico de linha, podemos visualizar melhor para analisar, e com isso, extrair algumas conclusões:

#### Conclusão 1: Os estados que começam com '1' no código são região Norte, então podemos pelo código, criar uma coluna na tabela com 'Região' de cada estado.

#### Conclusão 2: O estado do Pará tem mais que o dobro de internações comparado ao estado do Amazonas.

#### Conclusão 3: Percebe-se uma ligeira tendência de queda nos números de internações no estado do Pará ao passar dos meses.

#### Conclusão 4: Percebe-se uma ligeira tendência de subida nos números de internações no estado do Amazonas ao passar dos meses.

#### 4.2.3 Selecionando os cinco maiores estados por quantidade de internações.

Recriamos a coluna Total para ordenar os estados e selecionar os 5 maiores com a função **head**.

````python
dados['Total'] = dados.sum(axis= 1)
dados.head()
````

````python
ordenado_por_total = dados.sort_values('Total', ascending=False)
ordenado_por_total = ordenado_por_total.drop('Total', axis=1)
ordenado_por_total.head()
dados = dados.drop('Total', axis=1)
````

````python
ax = ordenado_por_total.head().T.plot(figsize=(10,6), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.legend(bbox_to_anchor = (1, 1))
plt.title("Os 5 Maiores Estados por números de internações")
plt.show()
````

#### Conclusão: No gráfico acima, verificamos subidas e quedas semelhantes nos maiores estados em números de internações. Principalmente com fortes quedas nos 5 estados nos meses mais recentes.

#### 4.2.4 Comparando os estados separando por Regiões.

````python
def RetornarRegiao(x):
    regiao = ''
    
    x = x[0]
    
    if (x == '1'): regiao = 'Norte'       
    elif (x == '2'): regiao = 'Nordeste'
    elif (x == '3'): regiao = 'Sudeste'
    elif (x == '4'): regiao = 'Sul'
    elif (x == '5'): regiao = 'Centro-Oeste'
    
    return regiao
````

````python
dados['Região'] = dados.index.map(RetornarRegiao)
dados.head()
````

````python
ax = dados[dados['Região'] == 'Nordeste'].drop('Região',axis=1).T.plot(figsize=(10,7.5), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Os Estados do Nordeste por números de internações")
ax.legend(bbox_to_anchor = (1, 1))
plt.show()
````

````python
ax = dados[dados['Região'] == 'Sudeste'].drop('Região',axis=1).T.plot(figsize=(10,7.5), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Os Estados do Sudeste por números de internações")
ax.legend(bbox_to_anchor = (1, 1))
plt.show()
````

````python
ax = dados[dados['Região'] == 'Sul'].drop('Região',axis=1).T.plot(figsize=(10,7.5), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Os Estados do Sul por números de internações")
ax.legend(bbox_to_anchor = (1, 1))
plt.show()
````

````python
ax = dados[dados['Região'] == 'Centro-Oeste'].drop('Região',axis=1).T.plot(figsize=(10,7.5), grid=True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Os Estados do Centro Oeste por números de internações")
ax.legend(bbox_to_anchor = (1, 1))
plt.show()
````

#### Conclusão: Nos gráficos de todas regiões acima, verificamos fortes quedas nos meses mais recentes. O que me leva a criar uma hipotese de que os dados mais recentes ainda não foram atualizados, podendo sofrer alterações futuramente.
